```python
import gensim
import pandas as pd
from gensim.utils import simple_preprocess
from gensim.parsing.preprocessing import STOPWORDS
from nltk.stem import WordNetLemmatizer, SnowballStemmer
from nltk.stem.porter import *
import numpy as np
from gensim import corpora
from gensim.models import TfidfModel
from pprint import pprint
np.random.seed(2021)
import nltk
nltk.download('wordnet')
```

```python
def Load_Dataset():
    data = pd.read_csv('Eluvio_DS_Challenge.csv',error_bad_lines = False, encoding ='utf-8' )
    data_text = data['title']
    temp = data_text
    return(data_text)
```

```python
load_dataset = Load_Dataset()
print(load_dataset)
#Load_Dataset
```

```python
def lemmatize_stemming(txt):
    return stemmer.stem(WordNetLemmatizer().lemmatize(txt,pos = 'v'))

def preprocess(txt):
    result = []
    for token in gensim.utils.simple_preprocess(txt):
        if token not in gensim.parsing.preprocessing.STOPWORDS and len(token) > 3:
            result.append(lemmatize_stemming(token))
    return result
stemmer = SnowballStemmer('english')
```

```python
def Res_Preprocess(load_dataset):
    processed_txt = load_dataset.map(preprocess)
    return processed_txt
```

```python
preprocessed = Res_Preprocess(load_dataset)
def dictionary(preprocessed):
    dictionary = gensim.corpora.Dictionary(second)
    return dictionary
    
    
```

```python
dictionary_1 = dictionary(preprocessed) 
def bag_of_words(dictionary_1,preprocessed):
    dictionary_1.filter_extremes(no_below = 15,no_above = 0.5, keep_n = 100000)
    bow_corpus = [dictionary_1.doc2bow(doc) for doc in preprocessed]
    return bow_corpus
```

```python
bow = bag_of_words(dictionary_1,preprocessed)
def Tfidf(bow):
    tfidf = TfidfModel(bow)
    corpus_tfidf = tfidf[bow]
    return corpus_tfidf
```

```python
tfidf = Tfidf(bow)
def LDA_model_BOW(bow,dictionary_1):
    lda_model = gensim.models.LdaMulticore(bow, num_topics=15, id2word=dictionary_1, passes=2, workers=2)
    return lda_model
```

```python
lda_model_bow = LDA_model_BOW(bow,dictionary_1)
```

```python
def print_topics_LDA_BOW(lda_model_bow):
    for idx, topic in lda_model_bow.print_topics(-1):
            print('Topic: {} \nWords: {}'.format(idx, topic))
print_topics_LDA_BOW(lda_model_bow)
```

```python
def LDA_model_tfidf(tfidf,dictionary_1):
    lda_model_tfidf = gensim.models.LdaMulticore(tfidf, num_topics=15, id2word=dictionary_1, passes=2, workers=4)
    return lda_model_tfidf
```

```python
lda_model_tfidf_1 = LDA_model_tfidf(tfidf,dictionary_1)
```

```python
def print_topics_LDA_Tfidf(lda_model_tfidf_1):
    for idx, topic in lda_model_tfidf_1.print_topics(-1):
        print('Topic: {} Word: {}'.format(idx, topic))
print_topics_LDA_Tfidf(lda_model_tfidf_1)
```

```python
doc_1 = 'How a Pentagon deal became an identity crisis for Google'
doc_2 = 'I had a peanut butter sandwich for breakfast. I like to eat almonds, peanuts and walnuts. My neighbor got a little dog yesterday. Cats and dogs are mortal enemies You mustnâ€™t feed peanuts to your dog.'

def test(doc,dictionary_1):
    bow_vector_1 = dictionary_1.doc2bow(preprocess(doc))
    for index, score in sorted(lda_model_bow[bow_vector_1], key=lambda tup: -1*tup[1]):
        print("Score: {}\t Topic: {}".format(score, lda_model_bow.print_topic(index, 5)))
    
test(doc_1,dictionary_1)
print()
test(doc_2,dictionary_1)
```
